```python
data['Age']=pd.cut(data['Age'],bins=[18,25,40,60],labels=['Youngster','Adult','Old Age'])
data['Age']=data['Age'].astype('object')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(8,4))
sns.histplot(data['Age'],color='skyblue')
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Education'],color='lightgreen')
plt.title('Distribution of Education')
plt.xlabel('Education')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Usage'],color='coral')
plt.title('Distribution of Fitness')
plt.xlabel('Fitness')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Fitness'],color='gold')
plt.title('Distribution of Fitness')
plt.xlabel('Fitness')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Income'],color='violet')
plt.title('Distribution of Income')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(data['Miles'],color='dodgerblue')
plt.title('Distribution of Miles')
plt.xlabel('Miles')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.set(style='whitegrid')
plt.figure(figsize=(8,4))
sns.countplot(data=data,x='Product')
plt.title('Count of Product')
plt.xlabel('Product')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x='Gender')
plt.title('Count of Gender')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.countplot(data=data,x='MaritalStatus')
plt.title('Count of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Frequency')
plt.show()
```

```python
numd=data.select_dtypes(include=['float64','int64'])
correlation_matrix=numd.corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.scatterplot(data=data, x='Income', y='Miles', color='dodgerblue')
plt.title("Income vs. Miles")
plt.xlabel("Income")
plt.ylabel("Miles")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=data,x='Product',y='Miles')
plt.title('Miles by Product')
plt.xlabel('Product')
plt.ylabel('Miles')
plt.show()
```
